## Triggers

**A trigger in MySQL is a database object that automatically executes (fires) a set of SQL statements in response to certain events on a specific table.**

#### ðŸ”¹ Why Use Triggers?

- Enforce business rules automatically.
- Keep audit logs of changes.
- Validate or transform data before saving.
- Automatically update related records.

<hr>

#### ðŸ”¹ Syntax

```sql
DELIMETER $$
CREATE TRIGGER trigger_name
timing event ON table_name
FOR EACH ROW
BEGIN
    -- SQL statements go here;
END$$
```
DELIMETER ;

- *timing* â†’ BEFORE or AFTER
- *event* â†’ INSERT, UPDATE, or DELETE
- *FOR EACH ROW* â†’ The trigger runs for each row affected.

<hr>

#### **A real world example would be to update one table after inserting values in another table:**

The follwoing: 
- creates a trigger name 'upd_emp_dem'
- triggers it when an insertion into employee_salary is done
- inserts values into employee_demographics

```sql

DELIMITER $$
CREATE TRIGGER upd_emp_dem
    AFTER INSERT ON employee_salary
    FOR EACH ROW
BEGIN
    INSERT INTO employee_demographics (employee_id, first_name, last_name)
    VALUES (NEW.employee_id, NEW.first_name, NEW.last_name);
END $$
DELIMITER ;
```
