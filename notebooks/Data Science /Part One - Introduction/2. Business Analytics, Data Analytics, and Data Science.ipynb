# ðŸ“˜ Study Guide: Various Data-Related Disciplines  

This study guide is structured to support deeper theoretical understanding and applied reasoning for exam-level questions relating to Business Analytics, Business Intelligence, Data Science, and Predictive/Prescriptive Analytics.  

---

# 1. Foundations of Analytical Disciplines

## 1.1 Business Analytics  
Business Analytics focuses on interpreting organisational data to support decision-making.  
It includes:
- Investigating *why* events occurred.
- Evaluating business processes.
- Using both qualitative and quantitative evidence.

### Key Capabilities  
- Identifying performance weaknesses.  
- Interpreting behavioural and operational patterns.  
- Supporting strategic adjustments.  

---

## 1.2 Business Intelligence (BI)  
BI provides descriptive insights by transforming historical data into structured visual outputs.

### BI Emphasis  
- Summarization rather than explanation.  
- KPI tracking and monitoring.  
- Dashboard development and reporting.

### Types of Dashboards  
**Static Dashboards**  
- Updated periodically (weekly, monthly).  
- Suitable for stable, slow-changing data.

**Real-Time Dashboards**  
- Continuously updated using low-latency pipelines.  
- Necessary when operational decisions require immediate information (e.g., live queues, system monitoring).

---

## 1.3 Data Science  
Data Science encompasses the processes, tools, and techniques used to uncover insights and produce predictive models.

### Core Components  
- Exploratory Data Analysis (EDA)  
- Data Cleaning & Wrangling  
- Predictive modelling  
- Scenario modelling  
- Optimization models  

Data Science moves beyond understanding the past to estimating probable future outcomes.

---

# 2. Exploratory and Diagnostic Techniques

## 2.1 Exploratory Data Analysis (EDA)  
EDA is the preliminary investigation of a datasetâ€™s structure, distributions, and patterns.

### Objectives  
- Identify anomalies (e.g., outliers, missing values).  
- Understand variable relationships.  
- Assess modelling assumptions.  
- Guide feature engineering.

### Importance  
- Prevents model misinterpretation.  
- Reduces risk of overfitting.  
- Improves data-driven decision quality.

---

## 2.2 Data Cleaning & Wrangling  
Data must be prepared before analysis or modelling.

### Key Processes  
- Standardizing formats.  
- Resolving inconsistencies.  
- Handling missing or duplicated data.  
- Detecting and treating outliers.  
- Validating data accuracy.

### Analytical Impact  
Clean data leads to:
- Trustworthy metrics.  
- Reliable dashboards.  
- Improved model performance.  

---

## 2.3 Root Cause Analysis (RCA)  
RCA identifies *why* a performance issue occurred.

### Methodological Elements  
- Multivariate investigation (examining many interacting factors).  
- Temporal comparison (before/after event).  
- Cause-and-effect mapping.  
- Separation of symptoms from true causes.

### Application Contexts  
- Unexpected KPI drops.  
- Operational failures.  
- Financial anomalies.

---

# 3. Predictive and Prescriptive Analytical Approaches

## 3.1 Predictive Analytics  
Predictive Analytics estimates future outcomes based on historical patterns and variables.

### Common Applications  
- Sales forecasting.  
- Churn prediction.  
- Risk scoring.  
- Demand estimation.

### Influencing Factors  
- Data quality and stability.  
- Presence of seasonality.  
- Degree of concept drift (changes in behaviour over time).

---

## 3.2 Concept Drift  
Concept drift occurs when the underlying data distribution changes, reducing model accuracy.

### Causes  
- Market shifts.  
- Behavioural changes.  
- External disruptions.

### Effects  
- Model degradation.  
- Incorrect predictions.  
- Need for retraining and monitoring.

---

## 3.3 Scenario Modelling (What-If Analysis)  
Scenario modelling evaluates the outcomes of hypothetical future decisions.

### Components  
- Baseline assumptions (based on validated data).  
- Variable manipulation.  
- Sensitivity and risk evaluation.  
- Comparison of alternative strategies.

### Strategic Value  
- Supports long-term planning.  
- Helps organisations prepare for uncertainty.  
- Reveals the consequences of strategic choices.

---

## 3.4 Optimization Models  
Used when the goal is to minimize cost or maximize efficiency.

### Applications  
- Industrial operations (e.g., drilling optimisation).  
- Production planning.  
- Scheduling and resource allocation.

### Structure  
- Predictive component (forecast behaviour).  
- Optimization layer (select best decision based on constraints).

---

# 4. Experimental and Qualitative Methods

## 4.1 A/B Testing (Controlled Experiments)

### Purpose  
To measure performance differences between controlled variants of an experience or process.

### Requirements  
- Randomized allocation.  
- Clearly defined KPIs.  
- Sufficient sample size.  
- Statistical validation.

### Uses  
- Web optimisation.  
- Product feature evaluation.  
- Messaging effectiveness.

---

## 4.2 Qualitative Analytics  
Qualitative data adds nuance and context that numerical models may overlook.

### Contributions  
- Reveals motivations behind behaviours.  
- Provides insight during early-stage adoption where data is sparse.  
- Enhances interpretation of quantitative patterns.

### Integration with Quantitative Models  
Combining approaches results in richer, more reliable predictions and recommendations.

---

# 5. Advanced Analytical Domains

## 5.1 Digital Signal Processing (DSP)  
DSP focuses on transforming, cleaning, and analysing continuous signals.

### Characteristics of DSP-Suitable Data  
- Time-dependent signals.  
- Sensor or equipment output.  
- Patterns with periodic noise.  
- High-frequency fluctuations.

### Common DSP Techniques  
- Smoothing  
- Noise filtering  
- Fourier transforms  
- Frequency-domain manipulation  

DSP is essential for industrial monitoring, audio engineering, and IoT analytics.

---

# 6. Data Infrastructure & Systems Reliability

## 6.1 Data Pipeline Integrity  
Dashboards and analytical systems rely on robust data pipelines.

### Indicators of Pipeline Weakness  
- Metric inconsistencies.  
- Lagged or missing updates.  
- Conflicting KPI values.  
- Frequent reprocessing failures.

### Consequences  
- Poor decision quality.  
- Misleading reporting.  
- Loss of executive trust.  

---

# 7. Analytical Domains Across Time Orientations

| Analytical Category | Primary Focus | Typical Tools | Example Uses |
|---------------------|--------------|---------------|--------------|
| **Descriptive** | What happened? | Dashboards, summaries | Performance monitoring |
| **Diagnostic** | Why did it happen? | RCA, EDA | Investigating KPI changes |
| **Predictive** | What will happen? | Forecasting, ML models | Revenue, churn, fraud risk |
| **Prescriptive** | What should we do? | Scenario models, optimisation | Strategic planning |

Understanding these distinctions is essential for determining the appropriate analytical method in exam scenarios.

---

# 8. How to Study These Concepts for Exam Success

### 8.1 Prioritise Conceptual Relationships  
Know how:
- EDA leads into modelling,  
- Data cleaning affects dashboards,  
- Qualitative insights enhance predictions,  
- Scenario modelling complements forecasting.

### 8.2 Practise Scenario-Based Reasoning  
Many exam questions require choosing the *most appropriate* technique based on context.

### 8.3 Understand the Purpose of Each Method  
Ask:
- What problem does this technique solve?  
- When is it the correct tool?  
- What assumptions does it rely on?  

### 8.4 Compare and Contrast Disciplines  
Be able to differentiate:
- Predictive vs prescriptive  
- Static vs real-time dashboards  
- Diagnostic vs descriptive analytics  

### 8.5 Study Real Examples  
Consider operational settings such as:
- Call centre queue monitoring  
- Marketing campaign analysis  
- Industrial sensor data  
- Customer churn patterns
