# Data Analysis Glossary

<hr>

## 1. Churn
**Definition:** The rate at which customers stop using a product or service over a given period.  
**Importance:** Helps businesses track customer retention, identify at-risk customers, and improve services.

## 2. KPI (Key Performance Indicator)
**Definition:** A measurable value that indicates how effectively a company is achieving its business objectives.  
**Importance:** Provides a clear metric to measure success and guide decision-making.

## 3. Conversion Rate
**Definition:** The percentage of users who complete a desired action, such as making a purchase or signing up.  
**Importance:** Shows the effectiveness of marketing campaigns, website design, or product features.

## 4. Bounce Rate
**Definition:** The percentage of visitors who leave a website after viewing only one page.  
**Importance:** Indicates how engaging or relevant the website content is; high bounce rates may highlight UX or content issues.

## 5. Retention Rate
**Definition:** The percentage of customers who continue using a product over time.  
**Importance:** Demonstrates customer loyalty and the long-term value of a product or service.

## 6. Data Cleaning
**Definition:** The process of detecting and correcting errors or inconsistencies in data.  
**Importance:** Ensures accurate analysis, reliable insights, and better decision-making.

## 7. Data Wrangling / Data Preparation
**Definition:** The process of transforming raw data into a usable format for analysis.  
**Importance:** Prepares data for visualization, modeling, or reporting, saving time and improving quality.

## 8. Descriptive Analytics
**Definition:** Analysis that summarizes historical data to understand what happened.  
**Importance:** Provides foundational insights for reporting and performance tracking.

## 9. Diagnostic Analytics
**Definition:** Analysis aimed at understanding why something happened.  
**Importance:** Helps identify root causes of trends, problems, or anomalies in data.

## 10. Predictive Analytics
**Definition:** Analysis that uses historical data to predict future outcomes.  
**Importance:** Enables forecasting and proactive decision-making.

## 11. Prescriptive Analytics
**Definition:** Analysis that recommends actions to achieve desired outcomes.  
**Importance:** Supports strategic decision-making by suggesting optimal actions.

## 12. Exploratory Data Analysis (EDA)
**Definition:** The process of examining datasets to summarize their main characteristics, often with visual methods.  
**Importance:** Helps discover patterns, anomalies, or relationships that guide deeper analysis.

## 13. Data Visualization
**Definition:** The representation of data using charts, graphs, or plots.  
**Importance:** Makes complex data easier to understand and communicates insights effectively.

## 14. Outliers
**Definition:** Data points that differ significantly from other observations.  
**Importance:** Can indicate errors, anomalies, or unique trends worth investigating.

## 15. Correlation
**Definition:** A statistical measure that describes the relationship between two variables.  
**Importance:** Helps identify associations that can inform predictions or strategies.

## 16. Data Pipeline
**Definition:** A set of processes that move data from sources to storage and eventually to analysis tools.  
**Importance:** Ensures a consistent and automated flow of data for analysis.

## 17. Hypothesis Testing
**Definition:** A method to test if a certain assumption about a dataset is true or false using statistical tests.  
**Importance:** Supports data-driven conclusions and reduces reliance on guesswork.

## 18. Sampling
**Definition:** Selecting a subset of data from a larger population for analysis.  
**Importance:** Allows faster and cost-effective analysis while maintaining accuracy if done correctly.

## 19. Data Integrity
**Definition:** The accuracy and consistency of data over its lifecycle.  
**Importance:** Critical for reliable insights and informed decision-making.

## 20. Dashboard
**Definition:** A visual interface that displays key metrics and KPIs in real-time.  
**Importance:** Helps stakeholders monitor performance and make timely decisions.

## 21. ETL (Extract, Transform, Load)
**Definition:** The process of extracting data from sources, transforming it into a usable format, and loading it into a database or warehouse.  
**Importance:** Fundamental for building clean, structured datasets for analysis.

## 22. Anomaly Detection
**Definition:** The identification of unusual patterns or data points that donâ€™t conform to expected behavior.  
**Importance:** Helps detect fraud, errors, or operational issues early.

## 23. Big Data
**Definition:** Extremely large and complex datasets that require advanced tools for processing and analysis.  
**Importance:** Provides richer insights and context for better decision-making.

## 24. Machine Learning (ML)
**Definition:** Algorithms that allow computers to learn patterns from data and make predictions or decisions.  
**Importance:** Enables predictive and prescriptive analytics at scale.

## 25. Data Governance
**Definition:** Policies and procedures that ensure data is managed properly and securely.  
**Importance:** Maintains trust, compliance, and quality in data-driven processes.
