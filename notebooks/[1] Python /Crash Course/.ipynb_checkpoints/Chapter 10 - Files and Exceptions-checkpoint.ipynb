## Reading from a file 

The following is a simple example of reading and printing the contents of a file: 

```
from pathlib import Path

path = Path('pi_digits.txt')
contents = path.read_text()
print(contents)
```

```read_text``` returns the contents of the file in the form of a single string 

In the example above the file pi_digits.txt would need to be in the same directory that this python script is saved in, in order for it to be successfully read, otherwise, the exact path to that files directory would need to be provided.
