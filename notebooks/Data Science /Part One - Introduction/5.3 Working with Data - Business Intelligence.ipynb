## **Techniques for Working with Information in Business Intelligence**

<br>

Business Intelligence (BI) focuses on **transforming processed data into meaningful information** that explains **past and current business performance** and supports **data-driven decision-making**.

Unlike traditional data processing or big data analytics, BI is primarily concerned with **descriptive and diagnostic analytics**.

<br>

<span style="color: lightgreen;">Processed Data → Analysis → Insights → Business Decisions</span>

<br><hr style="display: flex; width: 50%; margin: auto;"><br>

## **1. Purpose of Business Intelligence**

After data has been pre-processed, BI techniques combine:
- **business knowledge**
- **data analysis skills**

to answer questions such as:
- What happened?
- When did it happen?
- How many units did we sell?
- Which region performed best?
- How does performance compare to targets?

BI is mainly **backward-looking** (historical analysis), but highly valuable for management.

---

## **2. Measuring Business Performance**

### **Performance Measurement Flow**
**Observations → Quantification → Measures → Metrics → KPIs**

This structured flow is essential and is central to BI.

---

### **2.1 Observations**
Observations are **raw business events or variables**, such as:
- sales transactions
- customer sign-ups
- product returns
- website visits

Observations **cannot yet be analysed mathematically** in their raw form.

---

### **2.2 Quantification**
**Quantification** is the process of converting observations into numbers.

Examples:
- counting number of sales
- summing revenue
- calculating averages

Quantification makes analysis possible.

---

### **2.3 Measures**
**Measures** are the accumulated or aggregated results of quantified observations.

Characteristics:
- Based on historical data
- Often simple descriptive statistics
- Stored in fact tables

Examples:
- total monthly sales
- average order value
- number of new customers

---

### **2.4 Metrics**
**Metric = Measure + Business Meaning**

A metric gives context to a measure so it becomes meaningful to the business.

Examples:
- monthly sales growth rate
- customer churn rate
- revenue per region

Metrics allow comparison across:
- time periods
- regions
- products
- departments

---

### **2.5 Key Performance Indicators (KPIs)**
**KPI = Metric + Business Objective**

KPIs are **strategic**, goal-oriented metrics that indicate whether the business is succeeding.

Examples:
- sales growth ≥ 10%
- customer retention ≥ 85%
- order fulfilment time ≤ 48 hours

KPIs are:
- limited in number
- aligned to strategy
- monitored regularly

---

### ✅ **Data Aggregation**
Combining large volumes of processed data into:
- daily
- weekly
- monthly
- yearly summaries

This enables trend analysis and performance monitoring.

---

### ✅ **Dimensional Analysis**
Analysing measures across dimensions such as:
- time
- product
- region
- customer segment

This enables:
- drill-down (high-level → detail)
- roll-up (detail → summary)
- slicing and dicing

---

### ✅ **Comparative Analysis**
Comparing performance:
- against previous periods
- against targets
- across departments or regions

This is crucial for understanding performance changes.

---

### ✅ **Reporting**
BI reports:
- are structured and formatted
- provide static or scheduled summaries
- support operational and management decision-making

Examples:
- sales reports
- financial reports
- operational summaries

---

### ✅ **Dashboarding**
BI dashboards:
- provide visual, interactive summaries
- track KPIs in near real-time
- allow quick decision-making

Dashboards use:
- charts
- tables
- indicators
- alerts

---

### ✅ **Data Warehousing**
BI typically relies on:
- centralized data warehouses
- cleaned and integrated data
- historical data storage

### ✅ **ETL / ELT Processes**
Ensuring data flows correctly:
- Extract data from sources
- Transform data into a consistent format
- Load data into BI systems

---

## **5. Communication & Decision Support (Final Outcome)**

The final BI techniques focus on **communicating insights**:
- storytelling with data
- visualisation best practices
- executive summaries

These ensure that metrics and KPIs lead to **action**, not just reports.
