## Variables 
Variable names can contain only letters, numbers, and underscores. They can start with a letter or an underscore, but not with a number.  \
For instance, you can call a variable: \
✅ ```message_1``` \
but not \
❌ ```1_message.```

<br>

✅ variable references -> object (value) \
❌ variables do not store values 

<hr>

## Data Types 
Because most programs define and gather some sort of data and then do something useful with it, it helps to classify different types of data. 

- ### Strings
A string is a series of characters. Anythin placed within single or double quotes in python is a string. 

<br>
<h4 style="text-align: center;">Useful String Methods:</h4>
<br>

- title()
- upper()
- lower()

When storing data it is useful to store it using lower(), as user data is not trustworthy when it comes to capitalization

<br>
<h4 style="text-align: center;">Python f-strings</h4>
<br>

    
f-strings (f stands for format) allow variables to be passed within strings with which python replaces the variable with the object / value associated with that variable. 

<br>
<h4 style="text-align: center;">Adding whitespae to strings with Tabs and Newlines</h4>
<br>

To add a tab to text we can use the following character combination: `\t`

To add a newline we can use the following character combination: `\n`

<br>
<h4 style="text-align: center;">Stripping whitespace in python</h4>
<br>

We can either strip whitespace to the left of a string. the right, or all the whitespace. 

- lstrip()
- rstrip()
- strip()

Note that the above string methods do not alter the variables object in place. To alter the variables object we need to asign the variable to the updated value. 

For e.g. \
`x = x.strip()`

````python
# title() method example

name = 'jake smith'
name.title()
````

````python
#  String method example

name.upper()
````

````python
#  String method example

name.lower()
````

````python
# f-string example

first_name = 'Sarah'
last_name = 'Leslie'
full_name = f'{first_name} {last_name}'
full_name
````

````python
# strip() example

my_text = '   hello   world !    '
my_text = my_text.strip()
my_text
````

### Useful Terminal shortcuts (macos): 
- ctrl + L = Clear Python Terminal

### Useful Terminal commands (macos):
- cd (by itself) = is used to go to the home directory 
- cd .. = is used to naviagte around directories where .. represents directory names
- cd - = is used to go back to the previous directory 
- pwd = print working directory shows the path to the current directory
- ls = represents list, which shows all nonhidden files that exist in the current directory
