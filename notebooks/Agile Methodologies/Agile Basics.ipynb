# 🧠 Agile Methodologies: A Foundational Guide for Software Engineers

---

## 🚀 What is Agile?

**Agile** is a set of principles and values for **software development and project management** that promotes:

- Iterative development ✅  
- Collaboration 🤝  
- Flexibility and adaptability 🔄  
- Delivering small, working increments of software frequently ⏱️

---

## 📜 The Agile Manifesto (2001)

The core values of the Agile Manifesto are:

> ### 🧭 "We value:"
1. **Individuals and interactions** over processes and tools  
2. **Working software** over comprehensive documentation  
3. **Customer collaboration** over contract negotiation  
4. **Responding to change** over following a plan

---

## 🛠️ Why Agile is Important in Tech Roles

### For a Software Engineer:

- Encourages **constant feedback**, improving code quality 🧪
- Promotes **cross-functional team collaboration** 👥
- Helps manage **changing requirements** without derailing the project 🔄
- Speeds up **product delivery** and updates 🚚

---

## 🔁 Key Agile Frameworks (That You’ll Encounter)

### 1. **Scrum**
- Most popular Agile framework
- Based on **sprints** (1–4 week cycles)
- Roles:  
  - **Product Owner**: defines product backlog  
  - **Scrum Master**: facilitates Agile process  
  - **Development Team**: builds the product  

📋 **Ceremonies**:  
- Sprint Planning  
- Daily Standups  
- Sprint Review  
- Sprint Retrospective

---

### 2. **Kanban**
- Visual workflow management tool 🧾
- Uses **Kanban boards** with columns like:  
  `To Do | In Progress | In Review | Done`
- Focuses on **continuous delivery** and limiting work in progress (WIP)

---

### 3. **Extreme Programming (XP)**
- Prioritizes **engineering practices**
- Examples:  
  - **Pair programming**  
  - **Test-driven development (TDD)**  
  - **Continuous integration**

---

## 🧩 Core Concepts Every Engineer Should Know

### ✅ User Stories
- A way to describe software features from the end-user's perspective.\
As a [user], I want to [action] so that [value/benefit].

## 🔁 Iteration / Sprint
Time-boxed development cycles focused on building shippable software

## 🪵 Product Backlog
A prioritized list of features, bugs, and tasks to be completed

## 📊 Burndown Chart
A graph that shows remaining work versus time—used to track sprint progress
