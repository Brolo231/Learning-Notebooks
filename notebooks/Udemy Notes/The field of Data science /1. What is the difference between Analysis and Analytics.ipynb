# Difference Between **Analysis** and **Analytics**

---

## **Analysis**

Analysis involves breaking a large dataset into smaller, manageable parts and examining each section individually — as well as understanding how these parts relate to one another.  
It focuses on **past events**, aiming to explain **how** or **why** something happened.

### **Example**
- *“Sales dropped last month — here are the reasons why.”*

---

## **Analytics**

Analytics focuses on exploring **future possibilities**.  
It applies logical, statistical, and computational techniques to the components obtained from analysis, searching for patterns and insights that help predict or guide future decisions.

Analytics includes two main branches:

- **Qualitative Analytics** → intuition + analysis  
- **Quantitative Analytics** → formulas + algorithms

### **Example**
- *“Based on customer trends, sales are likely to increase next month if we launch a discount.”* \
- Why is this an example of Analytics? ⤵️ \
	•	Customer trends → existing data pattern \
	•	Likely to increase → prediction \
	•	If we launch a discount → data-driven scenario

---

## **Time Focus Summary**

- **Analysis = Past**  
  Understanding what happened and why.

- **Analytics = Present & Future**  
  Identifying patterns and predicting outcomes.

---

## **Summary**

From the definitions above, we can clearly see:

- **Analysis ≠ Analytics**  
- **Data Analysis ≠ Data Analytics**  
- **Business Analysis ≠ Business Analytics**
