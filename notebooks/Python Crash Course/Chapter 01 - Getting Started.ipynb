## Strings
A string is a series of characters. Anything placed within single or double quotes in python is a string. 

<br>

### Useful String Methods:

- title()
- upper()
- lower()

When storing data it is useful to store it using lower(), as user data is not trustworthy when it comes to capitalization

<br>

### Python f-strings
    
f-strings (f stands for format) allow variables to be passed within strings with which python replaces the variable with the object / value associated with that variable. 

<br>

### Adding whitespae to strings with Tabs and Newlines

To add a tab to text we can use the following character combination: `\t`

To add a newline we can use the following character combination: `\n`

<br>

### Stripping whitespace in python

We can either strip whitespace to the left of a string. the right, or all the whitespace. 

- lstrip()
- rstrip()
- strip()

Note that the above string methods do not alter the variables object in place. To alter the variables object we need to asign the variable to the updated value. 

For e.g. \
`x = x.strip()`

### Removing Prefixes and suffixes

We can use the following methods: `removeprefix(prefix_to_remove_from_string)` \
`removesuffix(...)`

Remember, just like how strip() does not alter the object associated with the variable that we apply these methods to, the same principle applies with removeprefix().

## Numbers

### Integers 
You can add (+), subtract (-), multiply (*), and divide (/) integers in Python.

```
>>> 2 + 3
5
>>> 3 - 2
1
>>> 2 * 3
6
>>> 3 / 2
1.5
```

Python uses two multiplication symbols to represent exponents:

```
>>> 3 ** 2
9
>>> 3 ** 3
27
>>> 10 ** 6
1000000
```

Python supports the order of operations too, so you can use multiple operations in one expression. You can also use parentheses to modify the order of operations so Python can evaluate your expression in the order you specify. For example:

```
>>> 2 + 3*4
14
>>> (2 + 3) * 4
20
```

### floats 
Python calls any number with a decimal point a float. This term is used in most programming languages, and it refers to the fact that a decimal point can appear at any position in a number.

For the most part floats can be treated as normal:

```
>>> 0.1 + 0.1
0.2
```

However, be aware that you can sometimes get an arbitrary number of decimal places in your answer:

```
>>> 0.2 + 0.1
0.30000000000000004
>>> 3 * 0.1
0.30000000000000004
```

Python tries to find a way to represent the result as precisely as possible, which is sometimes difficult given how computers have to represent numbers internally.

### Underscores in Numbers 

When you’re writing long numbers, you can group digits using underscores to make large numbers more readable:

```
>>> universe_age = 14_000_000_000
>>> universe_age
14000000000 # Notice how python only prints the digits
```

### Constants 

A constant is a variable whose value stays the same throughout the life of a program. Python doesn’t have built-in constant types, but Python programmers use all capital letters to indicate a variable should be treated as a constant and never be changed:

```
MAX_CONNECTIONS = 5000
```
