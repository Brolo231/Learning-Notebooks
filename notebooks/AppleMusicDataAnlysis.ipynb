# Apple Music - Exploratory Data Analysis

<hr>

### Every year apple music and spotify release their own versions of their customer bases listening analytics. 

<hr> 

#### The analytics include: 
- Top artists
- Top songs
- Top genres
- Top artist and song comparisons

<hr>

As someone who listens to music daily, I thought it would be interesting if I could replicate this concept by creating my own data anlysis on my own apple music library using SQL on MySQL Workbench.

## How?

The apple music library is only accessible through a macbook or windows laptop with the apple music application installed. To access your apple music library dataset you can: 

1. Open Apple Music
2. Select File > Library > Export library
3. Save the file

The following is a simplified overview of what the xml file structure looks like:
