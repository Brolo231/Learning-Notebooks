# WHERE

<hr>

#### The **WHERE** clause is used to filter rows from a table that meet a specific condition.

#### Instead of selecting all rows, you can narrow down your results based on criteria like job title, salary, or department.

<hr>

**âœ… Basic Syntax of WHERE**

```SELECT column1, column2```\
```FROM table_name```\
```WHERE condition;```

# LIKE

<hr>

#### The LIKE clause is used in a WHERE condition to search for a specified pattern in a column (usually text or VARCHAR fields).

<hr>

**âœ… Syntax:** 

```SELECT column1, column2```\
```FROM table_name```\
```WHERE column_name LIKE 'pattern';```

<hr>

**ðŸ”¡ SQL Wildcards Used with `LIKE`**

| Wildcard | Meaning                                  |
|----------|------------------------------------------|
| `%`      | Any number of characters, including 0    |
| `_`      | Exactly one character                    |

<hr>

**Note:**

MySQL table values can be case-sensitive, depending on the columnâ€™s collation settings. \
To ensure case-insensitive matching, use the LOWER() function on both sides of the comparison.

- The following returns all occupations that contain the word 'manager' regardless if it is Manager, MaNaGeR, or Assistant Manager

```SELECT *```\
```FROM employee_salary```\
```WHERE LOWER(occupation) LIKE '%manager%';```
