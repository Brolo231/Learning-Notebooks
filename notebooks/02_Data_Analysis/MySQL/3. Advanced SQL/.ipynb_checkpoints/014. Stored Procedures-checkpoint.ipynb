# Stored Prodcedures 

<hr>

#### Stored procedures are way of storing your sql code so that you can reuse it. 

<hr> 

### âœ… Syntax 

```sql
CREATE PROCEDURE procedure_name()
    query_statements
;
```

Then we need to *call* the stored procdure to you use:

```sql
CALL procedure_name();
```

<hr>

When creating stored procedures, it is likely that you will need multiple queries to be stored in the procedure. In this case we can make use of ```DELIMITER```

```sql
DELIMITER $$
CREATE stored_procedure()
BEGIN 
    SELECT * 
    FROM employee_salary
    WHERE salary > 50000;
    
    SELECT * 
    FROM employee_salary
    WHERE salary > 10000;
END $$
DELIMITER ;
```

<hr>

We can also use a *parameter* together with stored procedures which acts as a variable when we pass an argument to the stored procedure when we call it: 

For the following we are: 
1. Creating a stored procedure that accepts a parameter emp_id_param
2. the parameter's data type has to be declared - INT
3. we pass the parameter variable in a condition in the query
4. when we call the stored procedure we pass our value (In this case we assume we know that we want employee id 1, we just want the salary for that specific person)

```sql
DELIMITER $$
CREATE stored_procedure(emp_id_param INT)
BEGIN 
    
    SELECT salary
    FROM employee_salary
    WHERE employee_id = emp_id_param;
    
END $$
DELIMITER ;
```

CALL stored_procedure(1);
