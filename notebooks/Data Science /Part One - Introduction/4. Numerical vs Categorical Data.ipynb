# Numerical Data vs Categorical Data

## 1. What Does “Type of Data” Mean?
A **data type** describes **what kind of information a variable represents** and **what you are allowed to do with it**.

In other words:
- Can you calculate with it?
- Can you order it?
- Is it just a label?

This is why distinguishing between **numerical** and **categorical** data is so important.

---

## 2. Numerical Data (Quantitative Data)

### Definition
**Numerical data** represents values that can be **measured or counted** and expressed as numbers.

These values have **real numeric meaning**, not just labels.

### Key Characteristics
- Represents **quantity**
- Arithmetic operations are meaningful
- Can be analysed using statistical formulas
- Supports mathematical comparisons

### Examples
- Age (18, 25, 42)
- Height (170 cm, 185 cm)
- Income (R12 000)
- Number of purchases (3, 10)
- Exam score (78%)

### Types of Numerical Data
#### a) Discrete Numerical Data
- Countable values
- Whole numbers only  
Examples: number of customers, number of items sold

#### b) Continuous Numerical Data
- Measured values
- Can have decimals  
Examples: weight, temperature, time

### What It Means in Practice
Because numerical data has magnitude:
- You can calculate averages and totals
- You can measure differences
- You can use regression, correlations, and forecasts

---

## 3. Categorical Data (Qualitative Data)

### Definition
**Categorical data** represents **qualities or characteristics**, not measurements.

Numbers (if used) act only as **labels**, not values.

### Key Characteristics
- Represents **groups or categories**
- Arithmetic makes no logical sense
- Focuses on classification, not calculation
- Often analysed using counts and proportions

### Examples
- Gender (Male, Female)
- Product type (Laptop, Phone, Tablet)
- Payment method (Card, Cash)
- Customer segment (Student, Professional)
- Country (South Africa, UK)

⚠ Important: Even if categories are coded with numbers (e.g., Male = 1, Female = 2), they are **not numerical values**.

### Types of Categorical Data
#### a) Nominal Data
- Categories with **no order**
Examples: blood type, eye colour, country

#### b) Ordinal Data
- Categories with a **logical order**
Examples:
- Education level (High School → Degree → Masters)
- Satisfaction (Low → Medium → High)

---

## 4. Key Differences at a Glance

| Aspect | Numerical Data | Categorical Data |
|------|----------------|------------------|
| Represents | Quantity | Group or label |
| Arithmetic allowed | Yes | No |
| Can be averaged | Yes | No |
| Ordering | Yes | Sometimes |
| Examples | Age, income | Gender, category |

---

## 5. Why This Difference Matters

Understanding the difference tells you:
- **how to process the data**
- **which analysis techniques to use**
- **which visualizations make sense**

Examples:
- Numerical → averages, regression, line charts
- Categorical → counts, proportions, bar charts

Using the wrong type leads to **incorrect analysis and misleading results**.

---

## 6. In Simple Terms
- **Numerical data answers “How much?” or “How many?”**
- **Categorical data answers “Which type?” or “Which group?”**
