# What is Data Modelling?

#### Data modelling is the first step in database design.

<hr>

It’s a high-level, abstract phase (conceptual design) where the goal is to describe:
- The data in the database (e.g., students, lecturers, courses).
- The relationships between data (e.g., lecturers teach courses).
- The constraints on the data (e.g., student numbers have 8 digits).

The result is a conceptual representation of the database that ensures all user requirements are captured before implementation.

<hr>

## Steps in Database Design

1. Data Modelling – Defines entities, relationships, and constraints conceptually.
2. Logical Design – Converts the conceptual model into a DBMS-specific structure.
3. Physical Design – Defines how data is actually stored, organized, and accessed.

Each step moves from abstract (conceptual) to detailed (physical).

<hr>

#### Database design progresses through four abstraction levels, from most general to most specific:

| **Level** | **Description** | **Key Points** |
|------------|------------------|----------------|
| **External Model** | The user’s view of the database | - Represents individual department views (e.g., HR vs Sales)<br>- Each department has its own external schema |
| **Conceptual Model** | The overall community view | - Shows the logical structure of the entire database<br>- Includes entities, relationships, constraints, and business rules<br>- Independent of hardware and DBMS |
| **Internal Model** | The database as seen by the DBMS | - Maps the conceptual model into DBMS structures<br>- Includes tables, keys, and constraints<br>- Examples: Relational, Network, Hierarchical models |
| **Physical Model** | The actual storage structure | - Focuses on performance, storage, and access methods<br>- Defines how data is stored on disk (e.g., indexing, compression, encryption) |
